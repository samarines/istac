```python
from istacpy.indicators.lite import indicators
import pandas as pd

# ¿Qué temas existen en la base de datos?
indicators.get_subjects()
```

```python
indicators.get_indicators(subject_code='091')
```

```python
indicador = indicators.get_indicator('TRIBUTO_IRPF_RENTA_BRUTA_DECLARANTE')
indicador.info()
```

## Renta bruta por municipios

```python
dime = indicador.get_data(geo='M')
df_municipios = dime.as_dataframe()
df_municipios
```

```python
df_municipios.columns
```

## Renta bruta media de Arafo

```python
arafo = df_municipios.Arafo
arafo
```

## Renta bruta media de Santa Cruz de Tenerife

```python
santa_cruz = df_municipios['Santa Cruz de Tenerife']
santa_cruz
```

# Comparar las renta brutas medias de Arafo y Santa Cruz de Tenerife

```python
df_comparador = pd.concat([arafo, santa_cruz], axis=1)
df_comparador
```

```python
import matplotlib.pyplot as plt

x = df_comparador.index
y = df_comparador['Arafo']
z = df_comparador['Santa Cruz de Tenerife']


plt.figure(figsize=(10,6), dpi=100)
plt.plot(x, y, color='red', label='Arafo')
plt.plot(x, z, color='blue', label= 'Santa Cruz de Tenerife')
plt.xlabel('Años\n\nLa renta bruta media está compuesta por la suma de la renta bruta sujeta y de la renta bruta exenta dividida por el número de declaraciones del IRPF.')
plt.ylabel('Renta bruta en euros')
plt.legend(loc="upper left")
plt.title('Renta Bruta por declaración')
plt.grid()
plt.show()
```
